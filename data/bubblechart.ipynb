```
def fun(x):
  dat=x.copy()
  b=[]
  row=np.shape(x)[0]
  col=np.shape(x)[1]
  for i in range(row):
    if i==0:
      a=[]
      for j in range(col):
        a.append(dat[i,j])
        if j!=col-1:
          avg=(dat[i,j+1]+dat[i,j])/2
          a.append(avg)
      b=a
    else:
      a=[]
      for j in range(col):
        a.append(dat[i,j])
        if j!=col-1:
          avg=(dat[i,j+1]+dat[i,j])/2
          a.append(avg)
      b=np.vstack((b,a))
  return b

nconfirmed=fun(confirmed)

nma=fun(ma)

nfatality=fun(fatality)

ndna=fun(dna)
```
